# DATA PRE-PROCESSSING

### Since there aren't any significant correlation among the columns, we cannot delete any of the features

```python
num_columns=data[['Age','Years at Company','Monthly Income','Number of Promotions','Distance from Home','Number of Dependents','Company Tenure']]
plt.figure(figsize=(7,7))
sns.heatmap(num_columns.corr(),annot=True)
```

```python
### Information about the catgeorical features
data.describe(include='object')
```

### From the table above we can see that every object columns have less than or equal to 5 categories. So for easeness, we can first use Label encoding to convert these values to numbers
### Also, in the Attrition column, it is clear that the no. of entries for "Stayed" is 39113. Since there are no missing values, the no. of entries for "Left" will be 74461-39113=35348. So the weight distribution among these 2 values are 52.5% and 47.5% respectively.

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

object_col=data[['Gender','Job Role','Work-Life Balance','Job Satisfaction','Performance Rating','Overtime','Education Level','Marital Status','Job Level','Company Size','Remote Work','Leadership Opportunities','Innovation Opportunities','Company Reputation','Employee Recognition','Attrition']]

for i in object_col:
    data[i]=le.fit_transform(data[i])

data
```

### SCALING THE DATASET

```python
from sklearn.preprocessing import MinMaxScaler
mm=MinMaxScaler()
```

### Scaling all the numerical columns (Since the categorical columns have been encoded scall integers, we can avoid them)

```python
scaled=mm.fit_transform(num_columns)
scaled_num=pd.DataFrame(scaled)
scaled_num.columns=num_columns.columns
scaled_num
```

### Joining the scaled numerical columns with the label encoded features

```python
final=pd.concat([scaled_num,data[['Gender','Job Role','Work-Life Balance','Job Satisfaction','Performance Rating','Overtime','Education Level','Marital Status','Job Level','Company Size','Remote Work','Leadership Opportunities','Innovation Opportunities','Company Reputation','Employee Recognition','Attrition']]],join='outer',axis=1)
final
```

### Assigning the dependent (Attrition) and independent variables

```python
y=final['Attrition']
x=final.drop('Attrition',axis=1)
```

### Splitting the dataset into training and testing

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=60,test_size=0.2)
```

## RANDOM FOREST CLASSIFICATION
